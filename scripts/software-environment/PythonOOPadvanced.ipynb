```python
### Classmethod example
# "Classmethod is the alternative constructor. "

class Employee:
    
    num_of_emps = 0 ### Each time a new employee is cleated, it is increment by 1. -> init method
    raise_amt = 1.04
    
    def __init__(self, first, last, pay):
        self.first = first
        self.last = last
        self.pay = pay
        self.email = first + '.' + last + '@company.com'
        
        Employee.num_of_emps += 1
        ### This should be definitely the class variable method but not be the instance variable method "self"...
    
    def fullname(self): 
        return '{} {}'.format(self.first, self.last)

    def apply_raise(self): 
        self.pay = int(self.pay * self.raise_amt) #################
        
    @classmethod ### Decolator
    # This receives a class as the first argument. 
    def set_raise_amt(cls, amount):
        cls.raise_amt = amount
        
    @classmethod ### New Alternative Constructor
    def from_string(cls, emp_str): ### "from_" as convention. 
        first, last, pay = emp_str.split('-') 
        return cls(first, last, pay) # This line creates a new Employee

emp_1 = Employee("Tats", "SH", 50000) # This is going to set all attributes 
emp_2 = Employee("Test","User",60000) # This is going to set all attributes 

################# Parsing
emp_str_1 = 'John-Doe-70000'
emp_str_2 = 'Steve-AK-30000'
emp_str_3 = 'Jane-BBI-90000'

first, last, pay = emp_str_1.split('-')

new_emp_1 = Employee.from_string(emp_str_1) ### classmethod

print(new_emp_1.email)
print(new_emp_1.pay)

#####
# Use classmethod for alternative constructor, this is what that means. 

####
# Constructor. 
```

```python
### staticmethod example
# "Static method does not pass anything... "

class Employee:
    
    num_of_emps = 0 ### Each time a new employee is cleated, it is increment by 1. -> init method
    raise_amt = 1.04
    
    def __init__(self, first, last, pay):
        self.first = first
        self.last = last
        self.pay = pay
        self.email = first + '.' + last + '@company.com'
        
        Employee.num_of_emps += 1
        ### This should be definitely the class variable method but not be the instance variable method "self"...
    
    def fullname(self): 
        return '{} {}'.format(self.first, self.last)

    def apply_raise(self): 
        self.pay = int(self.pay * self.raise_amt) #################
        
    @classmethod ### Decolator
    # This receives a class as the first argument. 
    def set_raise_amt(cls, amount):
        cls.raise_amt = amount
        
    @classmethod ### New Alternative Constructor
    def from_string(cls, emp_str): ### "from_" as convention. 
        first, last, pay = emp_str.split('-') 
        return cls(first, last, pay) # This line creates a new Employee
    
    @staticmethod
    def is_workday(day): ### static method does not take class or instance as a first argument
        if day.weekday() == 5 or day.weekday() == 6:
            return False
        return True
    ### If you are not using self or cls, it is probably appropriate to use staticmethod. 

emp_1 = Employee("Tats", "SH", 50000) # This is going to set all attributes 
emp_2 = Employee("Test","User",60000) # This is going to set all attributes 

import datetime

my_date = datetime.date(2016,7,11)
print(Employee.is_workday(my_date)) ## Pass in the date. 

#####################
### Next one will be inheritance and subclass
```

```python
### Let's create Manager class

class Employee:
    
    raise_amt = 1.04
    
    def __init__(self, first, last, pay):
        self.first = first
        self.last = last
        self.pay = pay
        self.email = first + '.' + last + '@company.com'
    
    def fullname(self): 
        return '{} {}'.format(self.first, self.last)

    def apply_raise(self): 
        self.pay = int(self.pay * self.raise_amt) #################
        
class Developer(Employee):
    raise_amt = 1.10
    
    def __init__(self, first, last, pay, prog_lang):
        #super().__init__(first, last, pay)
        ###################################
        # Alternative method
        Employee.__init__(self, first, last, pay)
        self.prog_lang = prog_lang
        
class Manager(Employee):
    raise_amt = 1.10
    
    def __init__(self, first, last, pay, employees = None):
        #super().__init__(first, last, pay)
        ###################################
        # Alternative method
        Employee.__init__(self, first, last, pay)
        if employees is None:
            self.employees = []
        else:
            self.employees = employees
            
    def add_emp(self, emp):
        if emp not in self.employees: ########### non
            self.employees.append(emp)
            
    def remove_emp(self, emp):
        if emp in self.employees:
            self.employees.remove(emp)
            
    def print_emps(self):
        for emp in self.employees:
            print('-->', emp.fullname())


dev_1 = Developer("Tats", "SH", 50000, 'Python') # This is going to set all attributes 
dev_2 = Developer("Test","User",60000, 'Java') # This is going to set all attributes 

mgr_1 = Manager('Sue', 'SHEIS', 90000, [dev_1])

print(mgr_1.email)
print("*-----------------------------*")
mgr_1.print_emps()
mgr_1.add_emp(dev_2)
print("*-----------------------------*")
mgr_1.print_emps()
mgr_1.remove_emp(dev_1)
print("*-----------------------------*")
mgr_1.print_emps()

```

```python
# setter
class Employee6:

    def __init__(self, first, last):
        self.first = first
        self.last = last
        #self.email = first + '.' + last + '@email.com'

    @property
    def email(self):
        return '{}.{}@email.com'.format(self.first, self.last)

    @property
    def fullname(self):
        return '{} {}'.format(self.first, self.last)
    
    @fullname.setter
    def fullname(self, name):
        first, last = name.split(' ')
        self.first = first
        self.last = last


emp_1 = Employee6("Tats", "SH")

emp_1.first = 'Jim'
emp_1.fullname = 'Att Upt'


print(emp_1.first) # Changed
print(emp_1.email) # Changed 
print(emp_1.fullname) # Changed ()!
```

```python
# deleter
class Employee6:

    def __init__(self, first, last):
        self.first = first
        self.last = last
        #self.email = first + '.' + last + '@email.com'

    @property
    def email(self):
        return '{}.{}@email.com'.format(self.first, self.last)

    @property
    def fullname(self):
        return '{} {}'.format(self.first, self.last)
    
    @fullname.setter
    def fullname(self, name):
        first, last = name.split(' ')
        self.first = first
        self.last = last
        
    @fullname.deleter
    def fullname(self):
        print('Delete Name!')
        self.first = None
        self.last = None


emp_1 = Employee6("Tats", "SH")

emp_1.first = 'Jim'
emp_1.fullname = 'Att Upt'

print(emp_1.first) # Changed
print(emp_1.email) # Changed 
print(emp_1.fullname) # Changed ()!

del emp_1.fullname

print("*-------------------------------------*")
print(emp_1.first) # Changed
print(emp_1.email) # Changed 
print(emp_1.fullname) # Changed ()!
```
